import matplotlib and numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

create array x ranging from 0-1 in 100 steps

```python
x = np.linspace(0, 1, 100)
```

defines function returns exp(x)

```python
def exp_function(x):
    return np.exp(x)
```

set y = exp(x)
plot x vs y
Label the x-axis as time
label y-axis as awesomeness

```python
y = exp_function(x)
plt.plot(x, y)
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.title('Plot of exp(x)')
plt.savefig('figure_cj2-7', bbox_inches="tight", dpi=600)
```
