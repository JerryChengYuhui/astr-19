import numpy and matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

pull 1000 random numbers distributed uniformly between [0,1]

```python
x = np.random.uniform(0, 1, 1000)
```

Histogram the random numbers into 100 bins  
define xlabel and ylabel and title  
save the figure

```python
plt.hist(x, bins=100)

plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram of 1000 Uniformly Distributed Random Numbers')

plt.savefig('figure for cj2-9', bbox_inches = 'tight', dpi=600)
```
