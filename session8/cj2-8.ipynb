import numpy and matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

create an array in range 0-1 in 100 steps

```python
x = np.linspace(0,1,100)
```

define two functions:  
input x and it will return the value of sin(x)  
input x the other one will return cos(x).

```python
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

set two figures  
first plot in the left is sin(x) vs x, named the panel:"sin(x)"  
second plot in the right is cos(x) vs x, named the panel:"cos(x)"  
save the figure as PDF  

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot(x, sin_function(x))
ax1.set_xlabel('value of x')
ax1.set_ylabel('value of sin(x)')
ax1.set_title('sin(x)')

ax2.plot(x, cos_function(x))
ax2.set_xlabel('value of x')
ax2.set_ylabel('value of cos(s)')
ax2.set_title('cos(x)')

plt.savefig('figure for cj2-8', bbox_inches = 'tight', dpi=600)
```
