import matplotlib and numpy

```python
import matplotlib.pyplot as plt
import numpy as np
```

set mu is 0, sigma is 1  
generate 1000 random numbers from distribution

```python
mu, sigma = 0, 1
random_numbers = np.random.normal(mu, sigma, 1000)
```

create histogram with 100 bins  
label the xlabel, ylabel, title  
save the figure

```python
plt.hist(random_numbers, bins=100)

plt.xlabel('Value')
plt.ylabel('Frequency')

plt.title('Histogram of 1000 Random Numbers from a Normal Distribution')

plt.savefig('figure for cj2-10', bbox_inches = 'tight', dpi=600)
```
